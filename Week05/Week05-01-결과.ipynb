# 데이터 살펴보기


## 데이터 내용 미리보기
- 앞부분 미리보기 : DataFrame객체.**`head(n) => 기본 5개`**
- 뒷부분 미리보기 : DataFrame객체.**`tail(n) => 기본 5개`**
- 샘플 읽기 : DataFrame객체.**`sample(n) => 기본 1개`**

    * UCI Machine Learning 저장소에서 제공하는 자동차연비(auto mpg) 데이터셋을 사용한다.


```python
# 판다스 라이브러리 임포트

import pandas as pd
```

```python
# ./data/auto-mpg.csv파일을 heaer=None옵션을 설정하고 읽어온다.
# 파일로부터 읽은 내용을 변수 df에 대입한다.

df=pd.read_csv('./data/auto-mpg.csv', header=None)
```

```python
# df 확인

df
```

```python
# df의 타입확인

type(df)
```

```python
# 데이터프레임의 컬럼명을 '
# 'mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name'으로 설정한다.

df.columns=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']
```

```python
# df 확인

df
```

```python
# 데이터프레임 df의 앞 5개의 데이터를 확인한다.

df.head()
```

```python
# 데이터프레임 df의 마지막 3개의 데이터를 확인한다.

df.tail(3)
```

```python
# 데이터프레임 df의 샘플 데이터를 확인한다.

df.sample(5)
```

## 데이터 요약 정보 확인하기
- DataFrame의 크기(행, 열)확인: DataFrame객체.**`shape`**

```python
# 데이터프레임 df의 행,열 정보를 확인한다.

df.shape
```

- DataFrame의 기본 정보 출력: DataFrame객체.**`info()`**

```python
# df의 기본요약 정보 출력
# non-null: 유효한 값, 모두 유효한 값이면 결측치가 없다

df.info()
```

- DataFrame 각 열의 자료형 확인: DataFrame객체.**`dtypes`**


```python
# df 각 열의 데이터타입 확인

df.dtypes
```

```python
# 'mpg'열의 데이터타입을 확인한다.

df['mpg'].dtypes
```

- DataFrame의 기술 통계 정보 요약: DataFrame객체.**`describe()`**
  * **`산술(숫자) 데이터를 갖는 열`**에 대한 주요 기술 통계 정보(개수, 평균, 표준편차, 최대값, 최소값, 중간값 등)를 요약하여 출력함
  * **`inlude='object'`**옵션: 문자 데이터를 갖는 열에 대한 기술통계정보(개수, 고유값개수, 최빈값, 최빈값 개수)를 요약하여 출력함
  * **`includ='all'`**옵션: 모든 열에 대한 기술통계정보를 요약하여 출력함

```python
# 데이터프레임 df(수치데이터컬럼)의 기술통계 정보를 요약한다.

df.describe()
```

```python
# 데이터프레임 df의 문자열 데이터 컬럼들의 기술 통계 정보를 요약한다. include='object'

df.describe(include='object')
```

```python
# 데이터프레임 df의 모든 컬럼의 기술통계 정보를 요약한다.

df.describe(include='all')
```
